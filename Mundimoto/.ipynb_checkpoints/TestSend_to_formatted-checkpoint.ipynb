```python
import pandas as pd
import numpy as np

bikez_df = pd.read_csv('landing/persistent/all_bikez_data.csv')
brands_df = pd.read_csv('landing/persistent/bikez_brands_data.csv')
mundimoto_df = pd.read_csv('landing/persistent/mundimoto2.csv')

# ---
# *** Preprocessing mundimoto_df ***
# ---

# concat model and version to form model
mundimoto_df["Model"] = mundimoto_df['model'].astype(str) +" "+ mundimoto_df["version"]
mundimoto_df.drop(['model', 'version'], axis=1, inplace=True)

# renaming columns
mundimoto_df.rename(columns = {'brand':'Brand', 
                               'year':'Year', 
                               'km':'Odometer', 
                               'sell_price':'Sell_Price', 
                               'purchase_price':'Purchase_Price'}, 
                    inplace = True)

# reordering columns
mundimoto_df = mundimoto_df[['id', 'Brand', 'Model', 'Year', 'Odometer', 'Sell_Price', 'Purchase_Price']]


# ---
# *** Preprocessing bikez_df ***
# ---

# dropping unnecessary brand
brands_df.drop(brands_df[brands_df['Brand'] == 'Tiger '].index, inplace = True)


# ---
# *** Preprocessing bikez_df ***
# ---

# keeping only most important attributes
bikez_df = bikez_df[['Model', 
                     'Year', 
                     'Category', 
                     'Rating', 
                     'Displacement', 
                     'Engine type', 
                     'Power', 
                     'Cooling system', 
                     'Gearbox', 
                     'Transmission type', 
                     'Front tire', 
                     'Front brakes', 
                     'Rear tire', 
                     'Rear brakes', 
                     'Dry weight', 
                     'Wheelbase', 
                     'Fuel capacity', 
                     'Torque', 
                     'Compression', 
                     'Bore x stroke', 
                     'Seat height', 
                     'Valves per cylinder', 
                     'Fuel system', 
                     'Fuel control', 
                     'Front suspension', 
                     'Rear suspension', 
                     'Frame type', 
                     'Color options', 
                     'Overall length', 
                     'Ground clearance', 
                     'Overall height', 
                     'Overall width']]

#
#bikez_df.to_csv('landing/persistent/dani.csv', sep=';')
```

```python
import re

#i=1
brand_list = []
for model in bikez_df['Model']:
    for brand in brands_df['Brand']:
        regexp = re.compile(r'^{}'.format(brand.lower().strip()))
        if regexp.search(model.lower()):
            brand_list.append(brand.strip())
            #print(i, 'matched')
            break
    #i+=1
    
brand_target_df = pd.DataFrame(brand_list, columns=['Brand'])
brand_target_df.to_csv('wtf.csv', index=False)
```
